Idea behind the project: 

I used to play Tennis professionally, but quit with the age of 17. My friend Lenny Hampel became a Pro with his highest Rank being No. 356 on the ATP Tour. Unlike for the Top players, there is no statistical data on players above No. 150 on the ATP Tour. 

So my motivation was to create a dataset on Lenny Hampel's games, visualize it. Further, I would interview him, to see if his own subjective view about his game would match his statistical analysis (which he had never gotten before). Additionally I compared his stats, to Austria's No. 1 and world No. 3 Player Dominic Thiem. 

So I posed 2 Hypothesis for this project:<br>
H1: Lenny Hampel is overconfident in his assumptions due to self bias.<br>
H2: Lenny Hampel's (No.359) stats would be worse than of Dominic Thiem (No. 3)

Lmitations arise as the data set had to be stripped down to only 31 matches, as the data turned out to be missing for many of Lenny Hampels hundred matches which he played during this period.
```python
from bs4 import BeautifulSoup
import requests 
import json
import bs4 as bs
import urllib.request
from urllib.request import Request, urlopen
import pandas as pd
import pprint
import os
```

First the Data was collected from a livescore platform matchstat.com (https://matchstat.com/tennis/player/Lenny%20Hampel. The data was collected from 2016 and 2017 on the surface 'Clay'. The reason was that there was the most data avaliable, unlike in the years 2019 and 2020 due to inconsistency of matches. The files always contained stats to Lenny Hampel and his opponents, so it had to extracted and cleaned.

```python
tennisstat2016 = []

for number in range(1, 52):

    filename = f"lenny/2016/lenny2016_match ({number}).json"

    with open(filename) as json_file:
        lennymatch = json.load(json_file)
        
    player = [item 
         for item in lennymatch["stats"] 
         if item["player_fullname"] == "Lenny Hampel"]

    tabele = pd.DataFrame.from_dict(player)

    tennisstat2016.append(tabele)

# --- after loop ---

year2016 = pd.concat(tennisstat2016)
year2016
```

```python
tennisstat2017 = []

for number in range(1, 52):

    filename = f"lenny/2016/lenny2016_match ({number}).json"

    with open(filename) as json_file:
        lennymatch = json.load(json_file)
        
    player = [item 
         for item in lennymatch["stats"] 
         if item["player_fullname"] == "Lenny Hampel"]

    tabele = pd.DataFrame.from_dict(player)

    tennisstat2017.append(tabele)

# --- after loop ---

year2017 = pd.concat(tennisstat2017)
year2017
```

```python
years = [year2016, year2017]
finalresultyears = pd.concat(years)

finalresultyears.to_csv('finalresultyears.csv')
finalresultyears
```

Then the Data had to be cleaned and reordered, as there was data missing for matches. Additionally some metrics were not really useful for a comparison, thus they were also deleted. 

Thus the dataset has been aquired. The data has undergone manipulation in form of cleaning and reordering and a meaningful dataset 'lennydataset.csv' could be created. 

```python
finalresultyears_withoutcolums = finalresultyears.drop(['match_stats_id', 'serve_games_won', 'team', 'serve_games_won', 'winners', 'unforced_errors', 'serve_speed_fastest', 'serve_speed_1st_avg', 'serve_speed_2nd_avg', 'net_approaches', 'odds', 'match', 'id'], axis=1)
```

```python
finalresultyears_withoutcolums.dropna(subset=['total_points'], inplace=True)

finalresultyears_withoutcolums.to_csv('lennydataset.csv')

finalresultyears_withoutcolums
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
```

```python
df = pd.read_csv('lennydataset.csv')

df.drop(['Unnamed: 0'], axis=1, inplace = True)
```

```python
#new df with new relevant statistics
 
df['Percentage_1st_serve'] =  df['serve_1st_total'] / df['serve_1st_attempts']
df['Percentage_1st_serve_won'] =  df['serve_1st_won'] / df['serve_1st_total']
df['Percentage_2nd_serve'] =  df['serve_2nd_total'] / df['serve_1st_attempts']
df['Percentage_2nd_serve_won'] =  df['serve_2nd_won'] / df['serve_2nd_total']
df['Percentage_return_points'] =  df['return_points_won'] / df['return_points_total']
df['Percentage_return_to_break_points'] =  df['return_break_points_total'] / df['return_points_total']
df['Percentage_break_points'] =  df['return_break_points_won'] / df['return_break_points_total']



df
```

```python
plt.plot(df.Percentage_1st_serve, marker='^', markersize='5', linewidth='0.1')
plt.plot(df.Percentage_1st_serve_won, marker='^', markersize='5', linewidth='0.1')

plt.xlabel('Matches')
plt.ylabel('Percentage')
plt.title('Percentage 1st serve in vs. 1st serve won')
```

A Boxplot Analysis was the most suitable for these measures, as the range between matches is often
realtively high and thus it is possible to see the correct distribution of the data. It can be seen, that
Lenny Hampel has a big inconsistency in his first serve percentage, return points won and break points
percentage. While the last two are strongly influenced by the opponents serve, the first serve
percentage should be more consistent.

```python
plt.boxplot(df.Percentage_1st_serve)
plt.title('First Serve Percentage')
```
